```python
from scipy.stats import expon
```

```python
df = []
for _ in range(10000):
    X = expon.rvs(scale=0.5, size=1000)
    df.append(len(X) / sum(X))
```

```python
def var(df):
    mean = sum(df) / len(df)
    var = 0
    for x in df:
        var += (x - mean) ** 2
    var /= len(df)
    return var // 0.000001 * 0.000001

var100 = var(df[:100])
var1000 = var(df[:1000])
var10000 = var(df)
```

```python
import matplotlib.pyplot as plt
from math import floor
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

bins = floor(1.72 * 100 ** (1 / 3))

ax[0, 0].hist(df[:500], bins)
ax[1, 0].boxplot(df[:500])

bins = floor(1.72 * 1000 ** (1 / 3))

ax[0, 1].hist(df, bins)
ax[1, 1].boxplot(df)

bins = floor(1.72 * 10000 ** (1 / 3))

ax[0, 2].hist(df, bins)
ax[1, 2].boxplot(df)

ax[0, 0].set_title('$n=100$')
ax[0, 1].set_title('$n=1000$')
ax[0, 2].set_title('$n=10000$')

ax[1, 0].set_xlabel('$\sigma^2=$'+str(var100))
ax[1, 1].set_xlabel('$\sigma^2=$'+str(var1000))
ax[1, 2].set_xlabel('$\sigma^2=$'+str(var10000))

plt.show()
```
